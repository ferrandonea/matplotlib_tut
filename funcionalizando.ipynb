```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator


# datos dummy

data = [
    [1,43],
    [2,45],
    [3,42],
    [4,39],
    [5,38],
    [6,46],
    [7,46],
    [8,41],
    [9,35],
    [10,31],
    [11,38],
    [12,49]
]

delay_by_month = pd.DataFrame(data, columns = ["Month", "ArrDelay"]) #
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96) #
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6) #


ax.grid(which="major", axis="x", color="#DAD8D7", alpha=0.5, zorder=1) #
ax.grid(which="major", axis="y", color="#DAD8D7", alpha=0.5, zorder=1) #


ax.set_xlabel('', fontsize=12, labelpad=10)
ax.xaxis.set_label_position("bottom")
ax.xaxis.set_major_formatter(lambda s, i : f'{s:,.0f}')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.xaxis.set_tick_params(pad=2, labelbottom=True, bottom=True, labelsize=12, labelrotation=0)
labels = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ax.set_xticks(delay_by_month['Month'], labels, rotation=90)  # Map integers numbers from the series to labels list


ax.set_ylabel('Atrasos (minutos)', fontsize=12, labelpad=10)
ax.yaxis.set_label_position("left")
ax.yaxis.set_major_formatter(lambda s, i : f'{s:,.0f}')
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_tick_params(pad=2, labeltop=False, labelbottom=True, bottom=False, labelsize=12)

# ACA HAGO EL CAMBIO
ax.bar_label(bar1, labels=[f'{e:,.1f}'.replace(".",",") for e in delay_by_month['ArrDelay']], padding=3, color='black', fontsize=12)

ax.spines[['top','right']].set_visible(False)
ax.spines[['left','bottom']].set_linewidth(1.1)

ax.plot([0.12, .9], [.98, .98], transform=fig.transFigure, clip_on=False, color='navy', linewidth=.6)
ax.add_patch(plt.Rectangle((0.12,.98), 0.04, -0.02, facecolor='navy', transform=fig.transFigure, clip_on=False, linewidth = 0))


ax.text(x=0.12, y=.93, s="Promedio atraso aviones", transform=fig.transFigure, ha='left', fontsize=14, weight='bold', alpha=.8)
ax.text(x=0.12, y=.90, s="Diferencia en minutos entre hora fija y la real en promedio mensual", transform=fig.transFigure, ha='left', fontsize=12, alpha=.8)

ax.text(x=0.1, y=0.12, s="Fuente: Kaggle - Airlines Delay - https://www.kaggle.com/datasets/giovamata/airlinedelaycauses", transform=fig.transFigure, ha='left', fontsize=10, alpha=.7)


plt.subplots_adjust(left=None, bottom=0.2, right=None, top=0.85, wspace=None, hspace=None)


fig.patch.set_facecolor('white')


average = delay_by_month['ArrDelay'].mean()
below_average = delay_by_month[delay_by_month['ArrDelay']<average]
above_average = delay_by_month[delay_by_month['ArrDelay']>=average]

colors_high = ["#ff5a5f", "#c81d25"] # Extreme colours of the high scale
colors_low = ["#2196f3","#bbdefb"] # Extreme colours of the low scale

cmap_low = mpl.colors.LinearSegmentedColormap.from_list("low_map", colors_low, N=256)
cmap_high = mpl.colors.LinearSegmentedColormap.from_list("high_map", colors_high, N=256)
norm_low = mpl.colors.Normalize(below_average['ArrDelay'].min(), average) # linearly normalizes data into the [0.0, 1.0] interval
norm_high = mpl.colors.Normalize(average, above_average['ArrDelay'].max())

bar1 = ax.bar(below_average['Month'], below_average['ArrDelay'], color=cmap_low(norm_low(below_average['ArrDelay'])), width=0.6, label='Sobre promedio', zorder=2)
bar2 = ax.bar(above_average['Month'], above_average['ArrDelay'], color=cmap_high(norm_high(above_average['ArrDelay'])), width=0.6, label='Bajo promedio', zorder=2)
plt.axhline(y=average, color = 'grey', linewidth=3)

ymin, ymax = ax.get_ylim()
y_pos = average/ymax + 0.03

#ACA TAMBIEN
ax.text(0.88, y_pos, f'Promedio = {average:.1f}'.replace(".",","), ha='right', va='center', transform=ax.transAxes, size=12, zorder=3)
plt.show()
```
