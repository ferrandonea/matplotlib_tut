MaxNLocator es una clase de la biblioteca de visualización de datos Matplotlib que se utiliza para configurar los intervalos y los límites de los ejes de un gráfico.

En particular, MaxNLocator se utiliza para establecer el número máximo de divisiones en un eje y forzar la inclusión de valores de inicio y fin en el rango de datos para asegurarse de que los límites del eje estén bien definidos.

Esto es útil porque en algunos casos, Matplotlib puede establecer un número de divisiones de manera predeterminada que no es óptimo para visualizar los datos de manera clara y precisa. Usando MaxNLocator, se pueden configurar manualmente los intervalos del eje y asegurarse de que se muestren todos los datos relevantes en el gráfico.

Por ejemplo, si quieres establecer el número máximo de divisiones en el eje y a 5, puedes hacerlo de la siguiente manera : ax.yaxis.set_major_locator(MaxNLocator(nbins=5))

MaxNLocator(integer=True) es un locator de tipo MaxNLocator que se configura para asegurar que los ticks en el eje y sean números enteros. En otras palabras, esta línea de código fuerza a Matplotlib a mostrar solo valores enteros en el eje y de tu gráfico.

MaxNLocator es una clase de Matplotlib que se utiliza para configurar los intervalos y los límites de los ejes de un gráfico. A continuación, se muestran algunos de los atributos más comunes de la clase MaxNLocator:

atributos:
* nbins: este atributo establece el número máximo de divisiones en el eje. Por defecto, se establece en MaxNLocator.DEFAULT_NBINS, que es igual a 9.
* steps: este atributo es una secuencia de los pasos a considerar al elegir los límites y los intervalos del eje. Por defecto, es igual a MaxNLocator.STEP_LOOKUP, que es una secuencia que contiene los siguientes valores: 1, 2, 2.5 y 5.
* integer: este atributo es un valor booleano que indica si se deben mostrar solo valores enteros en el eje. Por defecto, es igual a False.
* symmetric: este atributo es un valor booleano que indica si los límites del eje deben ser simétricos. Por defecto, es igual a False.
* prune: este atributo es un valor booleano que indica si se deben eliminar las etiquetas que se superponen. Por defecto, es igual a True.
* min_n_ticks: este atributo establece el número mínimo de divisiones en el eje. Por defecto, se establece en MaxNLocator.MIN_N_TICKS, que es igual a 2.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar_label.html

* **labels** : array-like, optional : A list of label texts, that should be displayed. If not given, the label texts will be the data values formatted with fmt.

* **fmt** : str or callable, default: '%g' : An unnamed %-style or {}-style format string for the label or a function to call with the value as the first argument. When fmt is a string and can be interpreted in both formats, %-style takes precedence over {}-style.

* **label_type** : {'edge', 'center'}, default: 'edge' : he label type. Possible values:
    * 'edge': label placed at the end-point of the bar segment, and the value displayed will be the position of that end-point.
