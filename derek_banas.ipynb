## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#esto es para hacer auto reload
%reload_ext autoreload
%autoreload 2
```

## Functional Plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2

plt.plot(x_1,y_1)
plt.title("Days Squared Chart")
plt.xlabel("Days")
plt.ylabel("Days Squared")
plt.show()
```

## Print Multiple Plots

```python
plt.subplot(1,2,1) #el 1 de 1 row y 2 columns
plt.plot(x_1,y_1,'r')
plt.show()
```

```python
plt.subplot(1,2,1) #el 1 de 1 row y 2 columns
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2) #el 2 de 1 row y 2 columns
plt.plot(x_1,y_1,'b')
plt.show()
```

## Figure Object

Figure es un objeto que contiene todos los elementos del plot, puede tener mltiples axis

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)  #figzise en pulgadas
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #left bottom, width y height en porcentaje
#([a,b,c,d]) donde (a,b) is the point in southwest corner of the rectangle which we create. c represents width and d represents height of the respective rectangle.
axes_1.set_xlabel("Days")
axes_1.set_ylabel("Days Squared")
axes_1.set_title("Days Squared Chart")
axes_1.plot(x_1,y_1,'r', label="x/x2")
axes_1.plot(y_1,x_1,'b', label="x2/x")
axes_1.legend(loc=0) 
# 0 es best location, 1 es upper right, 2 upper left, 3 lower left, 4 lower right
# se puede dat tupla de (x,y) desde el lower left
plt.show()
```

Hagamos un grfico dentro de otro grafico, o un axis en otro axis

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100) 
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) 
axes_1.set_xlabel("Days")
axes_1.set_ylabel("Days Squared")
axes_1.set_title("Days Squared Chart")
axes_1.plot(x_1,y_1,'r', label="x/x2")
axes_1.plot(y_1,x_1,'b', label="x2/x")
axes_1.legend(loc=0) 

axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel("Days")
axes_2.set_ylabel("Days Squared")
axes_2.set_title("Days Squared Chart")
axes_2.plot(x_1,y_1,'k')

plt.show()
```

## Subplots

```python
fig2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout() #esto evita overlaps

axes_2[1].set_title("Plot 2")
axes_2[1].set_xlabel("X")
axes_2[1].set_ylabel("X Squared")
axes_2[1].plot(x_1,y_1)
plt.show()
```

## Appearence Options

```python
# Default colors (b: blue, g: green, r: red, c: cyan, m: magenta,
# y: yellow, k: black, w: white)
# color="0.75" creates a 75% gray
# You can use hexcodes color="#eeefff"
# You can use color names found next like this color="burlywood"
# https://en.wikipedia.org/wiki/Web_colors
# alpha defines the percentage of opacity

# The default line width is 1, so to double it put in 2 and so forth

# There are many line styles 
# matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
# You can also provide a sample like '-.'

# Markers can mark your provided points on the graph
# https://matplotlib.org/3.3.0/api/markers_api.html
# You can change the markersize as well

# markerfacecolor changes the marker fill color
# markeredgecolor changes the marker stroke color
# markeredgewidth changes the markers stroke size

#lw = linewidth
#ls = linestyle

fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])

axes_3.plot(x_1, y_1, color="navy" , alpha=0.75,
            lw=2, ls="-.", marker="o", markersize=7,
            markerfacecolor="yellow", markeredgecolor="y", markeredgewidth=4)

plt.show()
```

Quiero hacer un zoom

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])

axes_3.plot(x_1, y_1, color="navy" , alpha=0.75,
            lw=2, ls="-.", marker="o", markersize=7,
            markerfacecolor="yellow", markeredgecolor="y", markeredgewidth=4)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

plt.show()
```

agrego grid

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])

axes_3.plot(x_1, y_1, color="navy" , alpha=0.75,
            lw=2, ls="-.", marker="o", markersize=7,
            markerfacecolor="yellow", markeredgecolor="y", markeredgewidth=4)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True, color='0.6', dashes=(5,2,1,2))
axes_3.set_facecolor("#FAEBD7")
plt.show()
```
