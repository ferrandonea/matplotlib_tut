```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as matplotlib
```

```python
# datos dummy

data = [
    [1,43],
    [2,45],
    [3,42],
    [4,39],
    [5,38],
    [6,46],
    [7,46],
    [8,41],
    [9,35],
    [10,31],
    [11,38],
    [12,49]
]
```

```python
#asumamos que esos datos son el % de aviones atrasados
```

```python
#%matplotlib inline
#import matplotlib_inline
#|matplotlib_inline.backend_inline.set_matplotlib_formats('svg')

delay_by_month = pd.DataFrame(data, columns = ["Month", "ArrDelay"])
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6, zorder=0)
```

```python
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6)
# crear ejes
ax.grid(which="major", axis="x", color="#DAD8D7", alpha=0.5, zorder=1)
ax.grid(which="major", axis="y", color="#DAD8D7", alpha=0.5, zorder=1)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6, zorder=2)
# crear ejes
ax.grid(which="major", axis="both", color="#DAD8D7", alpha=0.5, zorder=1) 
# zorder Set the zorder for the artist. Artists with lower zorder values are drawn first.
plt.show()
```

Los kwargs están en https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html



```python
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6)
ax.grid(which="major", axis="both", color="#DAD8D7", alpha=0.5, zorder=1) 

# formateamos el eje X y los ticks
ax.set_xlabel('Mes del año', fontsize=12, labelpad=10) 
# labelpad: Spacing in points from the Axes bounding box including ticks and tick labels. If None, the previous value is left as is.
ax.xaxis.set_label_position("bottom") # top o bottom
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6)
ax.grid(which="major", axis="both", color="#DAD8D7", alpha=0.5, zorder=1) 

# formateamos el eje X y los ticks
ax.set_xlabel('Mes del año', fontsize=12, labelpad=10) 
ax.xaxis.set_label_position("top") # top o bottom
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (13.33, 7.5), dpi=96)
bar1 = ax.bar(delay_by_month["Month"], delay_by_month["ArrDelay"], width=0.6)
ax.grid(which="major", axis="both", color="#DAD8D7", alpha=0.5, zorder=1) 

# formateamos el eje X y los ticks
ax.set_xlabel('Mes del año', fontsize=12, labelpad=10, loc="left") 
#loc -> {'left', 'center', 'right'}, default: rcParams["xaxis.labellocation"] (default: 'center')
ax.xaxis.set_label_position("top") # top o bottom
plt.show()
```
