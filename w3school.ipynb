```python
## dos gráficos en un gráfico

x = np.random.randint(1,100,15)
y = np.random.randint(1,100,15)

print (f"x1: {x}")
print (f"y1: {y}")
print ("")

plt.scatter(x,y)

x = np.random.randint(1,100,15)
y = np.random.randint(1,100,15)

print (f"x2: {x}")
print (f"y2: {y}")

plt.scatter(x,y)
```

```python
## colores

x = np.random.randint(1,100,15)
y = np.random.randint(1,100,15)

print (f"x1: {x}")
print (f"y1: {y}")
print ("")

plt.scatter(x,y, color="hotpink")

x = np.random.randint(1,100,15)
y = np.random.randint(1,100,15)

print (f"x2: {x}")
print (f"y2: {y}")

plt.scatter(x,y, color="#88C999")
```

```python
# color en cada punto

colors = np.array(["red",
                   "green",
                   "blue",
                   "yellow",
                   "pink",
                   "black",
                   "orange",
                   "purple"])

x = np.random.randint(1,100,8)
y = np.random.randint(1,100,8)

print (f"x2: {x}")
print (f"y2: {y}")

plt.scatter(x,y, color=colors)
```
